## Stationarity check

### Dickey fuller test is used to check the stationarity of a time series. In this test:
### H0 - Time series in not stationary
### H1 - Time series is stationary


```python
from statsmodels.tsa.stattools import adfuller
```

```python
res=adfuller(DF['en'])
res
```

```python
print("Test statistic = ", res[0])
print("p value = ", res[1])
```

### Since p-value > 0.05, we failed to reject the null hypothesis. Hence the time series is not stationary.



### Converting the time series into stationary time series.

### 1. Remove trend and seasonality with decomposition

```python
ts = DF[['en']]
#ts=ts['en']
ts=ts.reset_index()
```

```python
ts.index = pd.to_datetime(ts.index)
ts
```

```python
ts.set_index('index', inplace=True)
ts
```

```python
import statsmodels.api as sm
model = sm.tsa.seasonal_decompose(ts,period=7)
model.plot()
```

```python
model.resid.plot()
```

```python
after = adfuller(model.resid.dropna())
after
```

```python
x=model.resid.dropna()
```

### The series is now stationary.

```python
train=ts[:-20]
test=ts[-20:]
```

### In ARMA family models, for getting p (Auto-regressive order) we look at PACF plot, there is a cut-off at lag1 due to which it will be considered as the order of AR in ARMA. Similarly q(Moving Average order) will also be equal to 1 as the value of ACF at lag2 is significantly different from that at lag1. Therefore lag1 is also a cut-off point in ACF plot.

```python
model = SARIMAX(train, order=(2, 1, 2))  
fitted = model.fit(disp=-1)  

# Forecast
fc = fitted.forecast(20, alpha=0.02)  

# Make as pandas series
fc_series = pd.Series(fc, index=test.index)
# Plot
plt.figure(figsize=(12,5), dpi=100)
plt.plot(train, label='training')
plt.plot(test, label='actual')
plt.plot(fc_series, label='forecast')

plt.title('Forecast vs Actuals')
plt.legend(loc='upper left', fontsize=8)
```

```python
mape = np.mean(np.abs(fc.values - test.values)/np.abs(test.values))
rmse = np.mean((fc.values - test.values)**2)**.5
print("mape:",mape)
print("rmse:",rmse)
```

### we can see that the model does not perform very well for multistep out sample data

### from the decomposition we can see that there is a weekly seasonality and still some spikes in the residual, that may be because of some external factors, which we can take into account by using them as our exogenous variable

```python
ex_df = pd.read_csv('Exog_Campaign_eng.csv')
ex_df.head()
```

```python
exog=ex_df['Exog'].to_numpy()
```

### we will train a sarimax model for that and see if we get anyimprovements from using the two information.

### the seasonal order and the values of PDQ are based upon various trials and comparision of the models

```python
import statsmodels.api as sm
train=ts[:530]
test=ts[530:]
model=sm.tsa.statespace.SARIMAX(train,order=(1, 1, 1),seasonal_order=(1,1,1,7),exog=exog[:530])
results=model.fit()

fc=results.forecast(20,dynamic=True,exog=pd.DataFrame(exog[530:]))

# Make as pandas series
fc_series = pd.Series(fc)
# Plot
train.index=train.index.astype('datetime64[ns]')
test.index=test.index.astype('datetime64[ns]')
plt.figure(figsize=(12,5), dpi=100)
plt.plot(train, label='training')
plt.plot(test, label='actual')
plt.plot(fc_series, label='forecast')

plt.title('Forecast vs Actuals')
plt.legend(loc='upper left', fontsize=8)
```

```python
mape = np.mean(np.abs(fc.values - test.values)/np.abs(test.values))
rmse = np.mean((fc.values - test.values)**2)**.5
print("mape:",mape)
print("rmse:",rmse)
```
